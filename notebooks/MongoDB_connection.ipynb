# <font color='#BFD72F' size=6>**2. Set MongoDB Atlas connection**</font> <a class="anchor" id="2"></a>

```python
username = os.getenv("PROJECT_USERNAME")
password = os.getenv("PROJECT_PASSWORD")
```

```python
print(username)
```

```python
# Set MongoDB Atlas connection parameters
mongo_uri = f"mongodb+srv://{username}:{password}@cluster0.dtgbnim.mongodb.net/?appName=Cluster0" 
databaseBikes = "Bikes"
collection_name_Bikes = "SeoulBikeData"

databaseBooks = "Books"
collection_name_Books = "BooksData"
```

```python
# We will instantiate the client
client = pymongo.MongoClient(mongo_uri)
```

```python
# Here, we create a new database and collection
db1 = client[databaseBikes]
collection1 = db1[collection_name_Bikes]

db2 = client[databaseBooks]
collection2 = db2[collection_name_Books]
```

```python
client.list_database_names()
```
